## Key Insights: 

### Overall Monthly Trends

Monthly trends in solar installation decreases while the monthly trend in capacity added increases.

### Overall Yearly Trends

In the first 2 years of solar installations, the number of installations is high but capacity added is low. Compared to the next few years, where the number of installation is lower then the first 2 years but the capacity added is almost the same. Number of installations in 2011 and 2020 is almost the same but the capacity added on both years have a very stark difference. 2011, yielded lesser kW compared to 2011 even though the number of Solar Installation is almost the same. The high yield in 2020 is also evident in the following years.

### Overall Trend per State

Queensland has the highest number of installation in total while NSW come in 2nd. In the kW Capacity added however, NSW comes first followed by Queensland. The rest of the state has the same rank on both install and capacity.

### Conclusion:

With new technology, the efficacy and capacity of solar power installed are getting bigger and higher. This is shown in the graph where the number of installation is getting lesser in the recent years but the capacity added was increasing. Showing that there is higher kW capacity per install in recent years that there was before. The trend could be that an average household in recent years prefer to install higher capacity inverters and more solar panels due to higher needs in electricty propelled by new technologies such as electric vehicles and solar batteries.

In the overall trend per state, although Queensland has a highest number of installation, the capacity added is lesser than that of NSW. This could be that more people in NSW has opted to get higher kW inverters and more solar panels per install.
